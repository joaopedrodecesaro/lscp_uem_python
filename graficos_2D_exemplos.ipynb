```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_list = ['Temp1', 'Temp2']
num_temp = len(temp_list)


"""
TIMESCALE
"""
days_in_period = 10
ini_time = 0
end_time = 360
num_periods = int(np.round((end_time-ini_time)/days_in_period)+1)
t = np.linspace(ini_time, end_time,num_periods)


coil_inlet_temperature = t+10
```

A imagem ficou boa, agora vamos criar uma função para facilitar a criação de mais gráficos.

```python
hex_colorscheme = ['#709431', '#C75F89']
hex_linestyle = ['-','--']

# Define function to plot property over time for multiple heat exchangers
def plot_aggreg_property_over_time(property_name, y_data, y_label, num_temp, hex_colorscheme, hex_linestyle, t,
                                  max_value = None, min_value = None, tick = None):
    plt.figure(figsize=(4, 3))

    plt.plot(t, y_data, label='mylabel', color=hex_colorscheme[0],
             linestyle=hex_linestyle[0], linewidth=2)

    plt.rcParams['font.size'] = 11

    plt.xlabel('Time (days)')
    plt.ylabel(y_label)

    # Set x-axis limits to include the last day (360)
    plt.xlim(0, 360)
    plt.xticks(ticks=range(0, 361, 40))

    # Set y-axis limits with buffer
    if max_value == None:
        max_value = y_data.max()
        min_value = y_data.min()
        buffer_ratio = 0.2  # adjust as needed
        y_range = max_value - min_value
        max_value_buffered = max_value + buffer_ratio * y_range
        min_value_buffered = min_value - buffer_ratio * y_range if min_value > 0 else 0
        plt.ylim(min_value_buffered, max_value_buffered)
    else:
        plt.ylim(min_value, max_value)
        plt.yticks(ticks=np.arange(min_value, max_value + 0.0000001, tick))

    plt.show()
```

```python
#These graphs have set y_min, y_max and y ticks.
plot_aggreg_property_over_time('Temperature', coil_inlet_temperature,
                              "Temperature (K)",num_temp, hex_colorscheme, hex_linestyle, t,
                              max_value = 400, min_value = 0, tick = 40)
```

```python
# Define function to plot property over time for multiple heat exchangers
def plot_fluid_property_over_time(property_name, y_data, y_label, num_temp, hex_colorscheme, hex_linestyle, t, particle=0,
                                  max_value = None, min_value = None, tick = None):
    plt.figure(figsize=(4, 3))

    for temp_num in range(num_temp):
        if num_temp == 1:
            mylabel = None
        else:
            mylabel = f'Temp-{temp_num+1}'
        plt.plot(t, y_data[:,temp_num], label=mylabel, color=hex_colorscheme[temp_num],
                 linestyle=hex_linestyle[temp_num], linewidth=1.5)

    plt.xlabel('Time (days)', fontsize=13)
    plt.ylabel(y_label, fontsize=13)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)

    # Set x-axis limits to include the last day (360)
    plt.xlim(0, 360)
    plt.xticks(ticks=range(0, 361, 40))

    # Set y-axis limits with buffer
    if max_value == None:
        max_value = y_data.max()
        min_value = y_data.min()
        buffer_ratio = 0.2  # adjust as needed
        y_range = max_value - min_value
        max_value_buffered = max_value + buffer_ratio * y_range
        min_value_buffered = min_value - buffer_ratio * y_range if min_value > 0 else 0
        plt.ylim(min_value_buffered, max_value_buffered)
        plt.yticks(fontsize=11)
    else:
        plt.ylim(min_value, max_value)
        plt.yticks(ticks=np.arange(min_value, max_value + 0.0000001, tick),fontsize=11)

    if num_temp > 1:
        plt.legend(fontsize=11, loc='upper center', bbox_to_anchor=(0.5, -0.20), fancybox=True, shadow=True, ncol=4)

    plt.show()
```

```python
# Define function to plot property over time for multiple heat exchangers
def plot_fluid_property_over_time(property_name, y_data, y_label, num_temp, hex_colorscheme, hex_linestyle, t, particle=0,
                                  max_value = None, min_value = None, tick = None):
    plt.figure(figsize=(5, 4))

    for temp_num in range(num_temp):
        if num_temp == 1:
            mylabel = None
        else:
            mylabel = f'Temp-{temp_num+1}'
        plt.scatter(t, y_data[:,temp_num], label=mylabel, color=hex_colorscheme[temp_num], linewidth=1)

    plt.xlabel('Time (days)', fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Set x-axis limits to include the last day (360)
    plt.xlim(0, 360)
    plt.xticks(ticks=range(0, 361, 40))

    # Set y-axis limits with buffer
    if max_value == None:
        max_value = y_data.max()
        min_value = y_data.min()
        buffer_ratio = 0.2  # adjust as needed
        y_range = max_value - min_value
        max_value_buffered = max_value + buffer_ratio * y_range
        min_value_buffered = min_value - buffer_ratio * y_range if min_value > 0 else 0
        plt.ylim(min_value_buffered, max_value_buffered)
        plt.yticks(fontsize=14)
    else:
        plt.ylim(min_value, max_value)
        plt.yticks(ticks=np.arange(min_value, max_value + 0.0000001, tick),fontsize=15)

    if num_temp > 1:
        plt.legend(fontsize=12, loc='upper center', bbox_to_anchor=(0.5, -0.25), fancybox=True, shadow=True, ncol=4)

    plt.show()
```

```python
def plot_fluid_property_bars(property_name, y_data, y_label, num_temp, hex_colorscheme, t,
                             max_value=None, min_value=None, tick=None, bar_width=8):
    plt.figure(figsize=(8, 4))

    t = np.array(t)
    num_times = len(t)

    if y_data.ndim == 1:
        y_data = y_data[:, np.newaxis]

    group_width = bar_width * num_temp
    offsets = np.linspace(-group_width/2 + bar_width/2, group_width/2 - bar_width/2, num_temp)

    for i in range(num_temp):
        plt.bar(t + offsets[i], y_data[:, i],
                width=bar_width,
                color=hex_colorscheme[i],
                label=f'Temp-{i+1}',
                edgecolor='black',
                linewidth=0.5)

    plt.xlabel('Time (days)', fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.xticks(t, fontsize=10)
    plt.yticks(fontsize=10)

    if max_value is None:
        max_value = y_data.max()
        min_value = y_data.min()
        buffer_ratio = 0.15
        y_range = max_value - min_value
        max_value += buffer_ratio * y_range
        min_value -= buffer_ratio * y_range if min_value > 0 else 0
        plt.ylim(min_value, max_value)
    else:
        plt.ylim(min_value, max_value)
        plt.yticks(ticks=np.arange(min_value, max_value + 1e-8, tick), fontsize=10)

    if num_temp > 1:
        plt.legend(fontsize=10, loc='upper right', frameon=False)

    plt.title(f'{property_name.title()} vs. Time', fontsize=13, weight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
t = np.array([0, 60, 120, 180, 240, 300, 360])
y_data = np.random.rand(len(t), 2) * 100
colors = ['#1f77b4', '#ff7f0e']  # blue and orange

plot_fluid_property_bars('Viscosity', y_data, 'Viscosity (cP)', num_temp=2,
                         hex_colorscheme=colors, t=t)
```
